We will Clean Dataset , with 5 important Functions :

(1) Detecting Missing Values : We will Detect how many missing Values we have.

(2) Removing or Dropping Missing Values : Where the Data is missing more than 50%

(3) String Operations : Using String Operations , we can Strip Unwanted Chars from the dataset

(4) Imputation Missing Values : WE will be Imputing the Missing Values , with the help of mean, median and mode

(5) Dropping Duplicates : At last we will remove duplicate Values as it will increase the data Quality and also maintain it Integrity.

```python
df = pd.read_csv('/content/sample_data123 - uncleaned_dataset.csv')  #Importing Data by using this method. You can give any name instead of 'df' here.
df
```

```python
null_values = df.isnull().sum() / len(df) * 100 #Detecting Percentage of Missing Values
null_values
```

```python
df = df.drop('Unwanted Data', axis=1) #we will remove an column as the no of Null value is more than 50%
```

```python
df['Name'] = df['Name'].str.strip('\,.<>')   #String Operation to Normalize the Text and remove errors
df['Age'] = df['Age'].str.strip('\,.<>')
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median()) #Using (median) Imputation Method to Fill Empty Cells
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean()) #Using (Mean) Imputation Method to Fill Empty Cells
```

```python
df['Salary'] #The imputation is done successfully
```

```python
df['Department'] = df['Department'].fillna(df['Department'].mode()[0]) #Here We are Using Mode Imputation Method to Fill 'Department' Column. Mode, return most reccuring elemnts from the dataset.
```

```python
df['Phone No'] #I have Printed 'Phone No' columns so as to check the rows properly. You can see here the numbers are not formatted propely here.
```

```python
df['Phone No'] = df['Phone No'].str.replace(r'[^0-9]','',regex=True) #Then We will use string operation method to correct the 'Phone No' Columns
```

```python
df #As you can see all the columns are properly formatted
```

```python
df = df.dropna(subset=['Join Date', 'Phone No','Name']) #Then Finally Removing Null Values , that cannot be imputed.
```

```python
null_values = df.isnull().sum() / len(df) * 100 #checking the 'df' for Null values. Now you can see there are no null values present in datase now.
null_values
```

```python
df.drop_duplicates() #Method to Drop Duplicate Values.
```

```python
df.to_csv('Output.csv') #Converting cleaned data to Download File
```
